### Question 2 <a id="q2a"></a>

What features have the greater impact on the price ?

I choose to use a simple linear regression because it's one of the best method in term of interpretability (after the Subset Selection LASSO), according to the book ['An Introduction to Statistical Learning'](https://www.statlearning.com/) (page 25).

```python
df_lis.head()
```

```python
#we will do a linear regression and see the coefficient
#train the model
X = df_lis.drop('price', axis=1)
y = df_lis['price']

# Split data into training and test data, and fit a linear model
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=.30, random_state=42)
lm_model = make_pipeline(LinearRegression())

# If our model works, it should just fit our model to the data. Otherwise, it will let us know.
lm_model.fit(X_train, y_train)
```

```python
#test
y_test_preds = lm_model.predict(X_test)# Predictions here
r2_test = r2_score(y_test, y_test_preds) # Rsquared here
r2_test
```

```python
#coefficeint
coefficients = lm_model.named_steps['linearregression'].coef_

# sort it
sorted_indices = coefficients.argsort()[::-1]
sorted_coefficients = coefficients[sorted_indices]
sorted_features = X_train.columns[sorted_indices]

# display
print("Coefficients (Weights) in Magnitude Order:")
for feature, coef in zip(sorted_features, sorted_coefficients):
    print(f"{feature}: {coef:.4f}")
```
